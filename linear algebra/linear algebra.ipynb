```python
#matrix multiplication
A = np.array([[1,2],
              [3,4]])

B = np.array([[0,2],
              [3,2]])

C = np.dot(A,B)
print(C)
```

```python
#Element Wise multiplication: Hadamard product


A*B   
```

```python
x = np.array([-3,4])

lp2 = np.linalg.norm(x)
print(lp2)

lp1 = np.linalg.norm(x,ord=1)
print(lp1)

lpinf = np.linalg.norm(x,ord=np.inf)
print(lpinf)
```

```python
#determinats
A = np.array([[1,2],
              [3,4]])


np.linalg.det(A)
```

```python
#inverse
A = np.array([[1,2],
              [3,4]])
Ainv = np.linalg.inv(A)

print(Ainv)

          
pinv = np.linalg.pinv(A)
print(pinv)
```

```python
#solving system of equations
a = np.array([[2,3], [3,1]])
b = np.array([8,5])
np.linalg.solve(a,b)
```
