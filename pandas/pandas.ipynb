```python
titles = df.get('Film')
```

```python
freq_title = {}

for t in titles:
    length = len(t)
    
    try:
        freq_title[length] += 1
    except:
        freq_title[length] = 1

print(freq_title)
```

```python
X  = np.array(list(freq_title.keys()))
Y = np.array(list(freq_title.values()))
```

```python
plt.scatter(X,Y)
plt.xlabel("Length Title")
plt.ylabel("No of movies")
plt.show()
```
